- A **task** is an instance of the Reinforcement Learning problem.
- **Continuing tasks** are tasks that continue forever, without end.
- **Episodic tasks** are tasks with a well-defined starting and ending point.
  - In this case, we refer to a complete sequence of interaction, from start to finish, as an **episode**.
  - Episodic tasks come to an end whenever the agent reaches a **terminal state**.
  - In episodic tasks, we use $\mathcal{S}^+$ to refer to the set of all states, including terminal states.

In the literature, **return** and **discounted return** are often used interchangably. For an arbitrary time step $t$, both refer to $G_t \doteq R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \ldots = \displaystyle\sum_{k=0}^\infty \gamma^k R_{t+k+1}$, where $\gamma \in [0,1]$.

_In particular, when people refer to **return**, it is not necessarily the case that $\gamma = 1$, and when they refer to **discounted return**, it is not necessarily true that $\gamma < 1$_.

In the classic reinforcement learning task of [Pole Balancing](https://gym.openai.com/envs/CartPole-v0/), a cart is positioned on a frictionless track, and a pole is attached to the top of the cart. The objective is to keep the pole from falling over by moving the cart either left or right, and without falling off the track.

![Pole Balancing](https://miro.medium.com/max/600/1*oMSg2_mKguAGKy1C64UFlw.gif)

The agent applies a force of +1 or -1 to the cart at every time step. It is formulated as an episodic task, where the episode ends when:
- the pole falls more than 20.9 degrees from vertical,
- the cart moves more than 2.4 units from the center of the track, or
- when more than 200 time steps have elapsed.

The agent receives a reward of +1 for every time step, including the final step of the episode.


#### The discount rates that would encourage the agent to keep the pole balanced for as long as possible, are
- [x] 1.0

- [x] 0.5
